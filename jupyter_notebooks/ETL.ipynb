# Section 2 - The Transformation

```python
# Making a copy of the raw dataset to work with for cleaning and transformation
df_clean = df_raw.copy()
df_clean.head(10)
```

```python
# Checking for the total number of null values in each column of the cleaned dataset
df_clean.isnull().sum()
```

```python
# Double checking for any duplicate rows
df_clean.duplicated().sum()
```

```python
# Firstly I need to clean the Gender column
# I will drop all 'NaN' and 'Other' values in the Gender column
df_clean = df_clean[df_clean['Gender'].isin(['Male', 'Female'])]
df_clean
```

```python
# Checking the cleaned Gender column
print(df_clean['Gender'].value_counts())
print(df_clean['Gender'].unique())
```

```python
# Replace all missing values in the Sleep_Hours_Per_Night column
df_clean['Sleep_Hours_Per_Night'] = df_clean['Sleep_Hours_Per_Night'].fillna(df_clean['Sleep_Hours_Per_Night'].mean())
df_clean['Sleep_Hours_Per_Night'].value_counts()  # Check the unique values in the column after filling missing values 

# After viewing the table, I see a result that needs to rounded to 1 decimal place
df_clean['Sleep_Hours_Per_Night'] = df_clean['Sleep_Hours_Per_Night'].round(1)
df_clean['Sleep_Hours_Per_Night'].value_counts()
```

```python
# Do the same for the remaining numerical columns 
numeric_cols = [
    'Screen_Time_Hours',
    'Social_Media_Hours',
    'Physical_Activity_Hours_Per_Week',
    'Previous_Semester_GPA',
    'Productivity_Score',
    'Attendance_Percentage',
    'Assignments_Completed',
    'Class_Participation_Score',
    'Stress_Level',
    'Motivation_Level',
    'Previous_Semester_GPA'
    
]
for col in numeric_cols:
    df_clean[col] = df_clean[col].fillna(df_clean[col].mean()).round(1)

df_clean[numeric_cols].isnull().sum()  # Confirm their is no more missing values in the numerical columns
```

```python
# Dropping columns not needed for analysis
drop_columns = ['Student_ID', 'Age', 'Study_Hours_Per_Day', 'Internet_Quality', 'Part_Time_Job', 'AI_Tool_Usage_Hours_Per_Week']
df_clean = df_clean.drop(columns=drop_columns)
df_clean
```

With the data cleaned, I'll now begin feature engineering for the analysis.

```python
# Start with encoding the Gender column

df_clean['Gender_Encoded'] = df_clean['Gender'].map({'Female': 0, 'Male': 1})
df_clean
```

```python
# Encoding categories for 'Sleep_Hours_Per_Night'

def Sleep_Quality(Hours):
    if Hours <= 6.9:
        return 'Poor'
    elif Hours > 6.9 and Hours <= 8:
        return 'Healthy'
    else:
        return 'Excessive'

df_clean['Sleep_Hours_Per_Night'] = df_clean['Sleep_Hours_Per_Night'].apply(Sleep_Quality) 
df_clean.head(10)
```

```python
# Encoding Screen time hours
# Assumed that this value is daily, not weekly

def Daily_Screen_Time(Hours):
    if Hours <= 2:
        return 'Low'
    elif Hours > 2 and Hours <= 4:
        return 'Moderate'
    else:
        return 'High'

df_clean['Screen_Time_Hours'] = df_clean['Screen_Time_Hours'].apply(Daily_Screen_Time)
df_clean.head(10)
```

```python
# Encoding Social Media Hours
# Also assumed that this value is daily

def Daily_Social_Media_Time(Hours):
    if Hours <= 1:
        return 'Low'
    elif Hours > 1 and Hours <= 2:
        return 'Moderate'
    else:
        return 'High'  

df_clean['Social_Media_Hours'] = df_clean['Social_Media_Hours'].apply(Daily_Social_Media_Time)
df_clean.head(10)   
```

```python
# Encoding physical activity hours per week
# NHS recommends at least 2.5 hours of exercise per week
# I will use that as a benchmark for the categories

def Weekly_Physical_Activity(Hours):
    if Hours <= 2:
        return 'Underactive'
    elif Hours > 2 and Hours <= 3:
        return 'Moderately Active'
    else:
        return 'Very Active'

df_clean['Physical_Activity_Hours_Per_Week'] = df_clean['Physical_Activity_Hours_Per_Week'].apply(Weekly_Physical_Activity)
df_clean.head(10)   
```
