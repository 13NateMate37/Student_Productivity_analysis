# Import libraries and Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
```

# Section 1 - Load the clean dataset


```python
# Load the clean dataset
df_cleaned = pd.read_csv("Dataset/Cleaned/Student_Productivity_Cleaned.csv")
df_cleaned.head(10)
```

---

# Section 2 - Trialing the hypotheses
# Hypothesis 1 - Females outperform Males

```python
#First a boxplot for the mean and median w/ summary


plt.figure(figsize=(9, 6))
sns.boxplot(
    data=df_cleaned,
    x='Gender',
    y='Productivity_Score'
)
summary = df_cleaned.groupby("Gender")["Productivity_Score"].agg(["mean", "median"])
print(summary)
plt.title("Female VS Male Productivity Scores")
plt.xlabel("Gender")
plt.ylabel("Productivity Score")
plt.show()
```

Here shows that it is infact quite close and only at the extreme ends do females out perform males

```python
# Further insight with a histogram

plt.figure(figsize=(9, 6))
sns.histplot(
    data=df_cleaned,
    x='Productivity_Score',
    hue='Gender',
    kde=True,
    stat='density',
    common_norm=False,
    bins=30
)

plt.title("Female VS Male Productivity Scores")
plt.xlabel("Productivity Score")
plt.ylabel("Density")
plt.show()
```

I'm interested to see how these scores correlate with the respective gender's grade point average



```python
# Scatterplot to visualise if there is a correlation between the gender productivity score
# and its respective grade point average

plt.figure(figsize=(9, 6))
sns.scatterplot(
    data=df_cleaned,
    x='Productivity_Score',
    y='Previous_Semester_GPA',
    hue='Gender',
    alpha=0.9,
    s=60
)

plt.title("Producivity Score aginst GPA by Gender")
plt.xlabel("Productivity_Score")
plt.ylabel("GPA")
plt.show()
```

The dataset and visuals does not support that females innately outperform males but that males are more likely to be in the lower percentiles of academic productivity.

# Hypothesis 2 - 7-9hrs is optimal for increasing productivity

```python
# Hours of sleep against productivity score
# Side by side for productivity and GPA

plt.figure(figsize=(15, 9))
plt.subplot(1, 2, 1)
sns.boxplot(
    data=df_cleaned,
    x='Sleep_Hours_Per_Night',
    y='Productivity_Score' 
)
plt.title("Sleep tier affect on Productivity Scores")
plt.xlabel("Sleep_Hours_Per_Night")
plt.ylabel("Productivity_Score")

plt.subplot(1, 2, 2)
sns.boxplot(
    data=df_cleaned,
    x='Sleep_Hours_Per_Night',
    y='Previous_Semester_GPA'
)
plt.title("Sleep tier affect on GPA")
plt.xlabel("Sleep_Hours_Per_Night")
plt.ylabel("Previous_Semester_GPA")

plt.tight_layout()
plt.show()
```

The healthy sleep tier does show to be beneficial to student productivity. Interestingly though, that does not seem to have a large affect on increasing ones GPA. Whereas the excessive sleep tier actually seems to have a lower bottom end as aposed to those in the poor sleep tier. Will explore a little further with another plot.

```python
# Same approach as above

plt.figure(figsize=(15, 9))
plt.subplot(1, 2, 1)
sns.histplot(
    data=df_cleaned,
    x='Productivity_Score',
    hue='Sleep_Hours_Per_Night',
    kde=True,
    stat='density',
    common_norm=False,
    bins=20
)

plt.title("Sleep tier affect on Productivity Scores")
plt.xlabel("Sleep_Hours_Per_Night")
plt.ylabel("Productivity_Score")

plt.subplot(1, 2, 2)
sns.histplot(
    data=df_cleaned,
    x='Previous_Semester_GPA',
    hue='Sleep_Hours_Per_Night',
    kde=True,
    stat='density',
    common_norm=False,
    bins=20
)

plt.title("Sleep tier affect on Previous_Semester_GPA")
plt.xlabel("Sleep_Hours_Per_Night")
plt.ylabel("Previous_Semester_GPA")
plt.show()
```

This is a more concise view, supporting my prior statement and my second hypothesis.

# Hypothesis 3 - Excessive screen time and social media consumption is detrimental to productivity

```python
# Screen time and social media consumption against productivity

plt.figure(figsize=(15, 9))

plt.subplot(1, 2, 1)
sns.scatterplot(data=df_cleaned, x='Screen_Time_Hours', y='Productivity_Score')
plt.title('Screen Time vs Productivity')
plt.xlabel('Screen Time')
plt.ylabel('Productivity Score')

plt.subplot(1, 2, 2)
sns.scatterplot(data=df_cleaned, x='Social_Media_Hours', y='Productivity_Score')
plt.title('Social Media vs Productivity')
plt.xlabel('Social Media Consumption')
plt.ylabel('Productivity Score')


plt.tight_layout()
plt.show()
```

It appears that high social media consumption does negatively impact prodcutivity despite the fact high screen time shows to boost productivity

```python
# Screen time and social media consumption against GPA

plt.subplot(1, 2, 1)
sns.scatterplot(data=df_cleaned, x='Screen_Time_Hours', y='Previous_Semester_GPA')
plt.title('Screen Time vs GPA')
plt.xlabel('Screen Time (hours)')
plt.ylabel('GPA')

plt.subplot(1, 2, 2)
sns.scatterplot(data=df_cleaned, x='Social_Media_Hours', y='Previous_Semester_GPA')
plt.title('Social Media vs GPA')
plt.xlabel('Social Media Consumption')
plt.ylabel('GPA')

plt.tight_layout()
plt.show()
```

Here we can see that low screen time has a tighter range than high screen time but doesn't benefit GPA as much. While the low social media consumption does show stronger groupings as the GPA value increases despite having a similar range.

# Hypothesis 4 - Physical activity is beneficial to productivity

```python
# Physical activity plotted against productivty and GPA

plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
sns.histplot(
    data=df_cleaned,
    x='Productivity_Score',
    hue='Physical_Activity_Hours_Per_Week',
    kde=True,
    stat='density',
    common_norm=False,
    bins=20
)
plt.title("Physical Activity affect on Productivity Scores")
plt.xlabel("Physical Activity Hours Per Week")
plt.ylabel("Productivity Score")

plt.subplot(1, 2, 2)
sns.histplot(
    data=df_cleaned,
    x='Previous_Semester_GPA',
    hue='Physical_Activity_Hours_Per_Week',
    kde=True,
    stat='density',
    common_norm=False,
    bins=20
)
plt.title("Physical Activity affect on Previous Semester GPA")
plt.xlabel("Physical Activity Hours Per Week")
plt.ylabel("Previous Semester GPA")

plt.tight_layout()
plt.show()
```

Here we can see that being underactive does reduce productivity but does not necessarily bring down the GPA. Very active students also seem to have a lower GPA despite having a higher productivty.
